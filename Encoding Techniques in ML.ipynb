```python
#Loop

for column in df.columns:
    if df[column].dtype ==np.number:
        continue
    df[column] = LabelEncoder().fit_transform(df[column])
```

# Ordinal Encoder

```python
df =pd.read_csv('agora.csv')
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.Area.unique()
```

```python
city_list=['Dhaka','Ctg','Rangpur']
```

```python
ordinal =OrdinalEncoder(categories=[city_list])
```

```python
#Its working number & samples
encoded_values=ordinal.fit_transform(df[['Area']])
```

```python
pd.DataFrame(encoded_values, columns=['Area'])
```

```python
new_area=pd.DataFrame(encoded_values, columns=['Area'])
```

```python
df.head()
```

```python
df.Area =new_area
```

```python
df.head()
```
